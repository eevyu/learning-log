### matplotlib 라이브러리
- 그래프를 그리고 싶을 때 사용
- 그래프마다 입력해주어야 할 내용이 조금씩 다름
- 그럼에도 '데이터'는 모든 그래프에 꼭 필요함
1. 라이브러리 import하기 : import matplotlib.pyplot as plt (as - 'alias', 별명)
2. 데이터를 적절한 그래프로 그리기 : plt.plot([10, -4, 15, 19]) (plot - 꺾은선 그래프)
3. 그래프 보여주기 : plt.show()

#### 기본 그래프 TOP 5!

```python
import matplotlib.pyplot as plt
```

##### 1) 꺾은선 그래프(plot) - y축 데이터만 있어도 됨

```python
plt.plot([-10, 5, 20, 13])
plt.show()
```

##### 2) 막대 그래프(bar, barh) - 수직, 수평

```python
plt.bar([1, 2, 3, 4], [-10, 5, 13, 3])
plt.show()
```

```python
plt.barh([1, 2, 3, 4], [-10, 5, 13, 3])
plt.show()
```

##### 3) 히스토그램(hist)

```python
data = [-10.2, -5.2, -8.0, -5.6, -7.8, -5.9, -5.9, -5.0, -1.2, -3.4, -10.3, -11.3]
plt.hist(data, bins = 15) # bins 생략 시 10구간으로 고정
plt.show()
```

##### 4) 박스플랏(boxplot) - 최소, 중앙(25%, 50%, 75%), 최댓값

```python
data = [-10.2, -5.2, -8.0, -5.6, -7.8, -5.9, -5.9, -5.0, -1.2, -3.4, -10.3, -11.3]
plt.boxplot(data)
plt.show()
```

##### 5)산점도(scatter)

```python
high = [2.3, 3.0, 2.5, 1.0, 1.9, 3.3, 4.5, 3.7, 5.6, 4.4]
low = [-10.2, -5.2, -8.0, -5.6, -7.8, -5.9, -5.9, -5.0, -1.2, -3.4]
plt.scatter(high, low)
plt.show()
```

#### 그래프 꾸미기

```python
# 한글 폰트 깨짐 해결 - 기본 폰트로 '맑은 고딕' 사용
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

plt.title('맑은 고딕 테스트')
plt.xlabel('최고기온')
plt.ylabel('최저기온')
plt.plot([1, 2, 3], [-10, 20, 30])
plt.show()
```

```python
high = [2.3, 3.0, 2.5, 1.0, 1.9, 3.3, 4.5, 3.7, 5.6, 4.4]
low = [-10.2, -5.2, -8.0, -5.6, -7.8, -5.9, -5.9, -5.0, -1.2, -3.4]

# 그래프 제목
plt.title('1월의 최고-최저기온 관계')

# x축, y축 이름
plt.xlabel('최고기온')
plt.ylabel('최저기온')

# 범례 만들기, 색 설정
plt.plot(low, c = 'skyblue', label = '1월 최저기온')
plt.plot(high, c = 'lightpink', label = '1월 최고기온')
plt.legend()
plt.show()
```

```python
plt.scatter(low, high, c = high)
plt.jet()
plt.colorbar()
plt.show()
```
