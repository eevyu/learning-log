##### 2) for 반복문으로 파일에서 한 줄씩 읽어오기
for A in B :
    반복할 명령

```python
# 명단 파일 한 줄씩 출력하기

with open("우리반명단.txt", mode='r', encoding='utf-8') as data: # UTF-8 인코딩으로 파일 열기
    for row in data:
        print(row)
```

```python
# 우리반 명단 리스트에 저장하기

names = []
with open("우리반명단.txt", mode='r', encoding='utf-8') as data:
    for row in data :
        names.append(row)
print(names)
```

```python
#랜덤하게 한 명 뽑기

import random
print(random.choice(names))
```

```python
# 두구두구두구!! 오늘의 주인공은!!

import time
print('오늘의 주인공은 바로!!')
time.sleep(3)
print(random.choice(names))
```

```python
# 명단 파일에서 궁금한 성씨를 가진 사람만 출력하기

with open("우리반명단.txt", mode='r', encoding='utf-8') as data:
    for row in data :
        if row[0] == '송' :
            print(row)
```

```python
# 명단 파일에서 궁금한 성씨를 가진 사람만 리스트에 추가하기

x = input('궁금한 성씨를 입력해주세요 : ')
names = []
with open("우리반명단.txt", mode='r', encoding='utf-8') as data:
    for row in data :
        if row[0] == x :
            names.append(row.strip())
print(names)
```

```python
# csv 데이터 읽어오기

import csv
data = csv.reader(open('seoultemp.csv'))
print(data)
```

```python
# 데이터 한 줄씩 읽어오기

data = csv.reader(open('seoultemp.csv'))
next(data)
for row in data :
    print(row)
```

```python
# 최고 기온만 출력해보기 (1950년대의 데이터가 비어있어 오류 발생)

data = csv.reader(open('seoultemp.csv'))
next(data)
for row in data :
    print(float(row[-1]))
```

```python
# 최고 기온 중 데이터가 비어있지 않은 날짜만 출력해보기

data = csv.reader(open('seoultemp.csv'))
next(data)
for row in data :
    if row[-1] != '' :
        print(row[0])
```

##### 4) 진짜 데이터로 몸을 풀어봅시다!

```python
# 최고 기온 그래프로 표현해보기

data = csv.reader(open('seoultemp.csv'))
next(data)
temp = []
for row in data :
    if row[-1] != '' :
        temp.append(float(row[-1]))

import matplotlib.pyplot as plt
plt.plot(temp)
plt.show()
plt.boxplot(temp)
plt.show()
plt.hist(temp, bins = 100)
plt.show()
```

```python
# 우리 지역의 역대 최고 기온은?

data = csv.reader(open('seoultemp.csv'))
next(data)
memo = -1000
date = ''
for row in data :
    if row[-1] != '' :
        if memo < float(row[-1]) :
            memo = float(row[-1])
            date = row[0]
print(date, memo)
```

```python
# 지난 100년 간, 내 생일의 기온을 그래프로 표현한다면?
# a in b : a가 b에 포함되었는가?

data = csv.reader(open('seoultemp.csv'))
next(data)
x = []
for row in data :
    if row[-1] != '' :
        if '-01-18' in row[0] :
            x.append(float(row[-1]))

plt.plot(x)
plt.show()
plt.boxplot(x)
plt.show()
```

```python
# 1년 중 일교차가 가장 큰 달은 몇 월일까요?

data = csv.reader(open('seoultemp.csv'))
next(data)
x = []
for i in range(12) :
    x.append([])
print(x)

for row in data :
    if row[-1] != '' and row[-2] != '' :
        x[int(row[0].split('-')[1]) - 1].append(float(row[-1]) - float(row[-2]))

plt.boxplot(x)
plt.show()
```

```python
# 1년 중 언제가 가장 덥고, 언제가 가장 추울까요?

data = csv.reader(open('seoultemp.csv'))
next(data)
temp = []
for i in range(31) :
    temp.append([])

for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '08' :
            temp[int(row[0].split('-')[2]) - 1].append(float(row[-1]))

plt.boxplot(temp)
plt.show()
```
